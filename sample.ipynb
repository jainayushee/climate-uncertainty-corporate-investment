```python
import pandas as pd

df = pd.read_csv("Fin-dataset.csv")
df.head()
```

```python
print(df.shape)
df.isnull().sum()
```

```python
print(len(df["conm"].unique()))
df["conm"].unique()
```

```python
grouped_data = df.groupby('GVKEY')

# Create an empty dictionary to store the separate data frames
company_dfs = {}

# Iterate over the groups and create a separate data frame for each
for company_id, group_df in grouped_data:
    # Store each group in the dictionary with the company ID as the key
    company_dfs[company_id] = group_df
```

```python
df["liquidity"] = df["at"]/df["lct"]
df["leverage"] = df["lct"]/df["ceq"]
df["ROA"] = df["ni"]/df["at"]

df.head()
```

```python
df_sorted = df.sort_values(by=['conm', 'fyear'])

# Calculate the change in assets for each year
df_sorted['Asset Change'] = df_sorted.groupby('conm')['at'].diff()

# Calculate the growth rate
df_sorted['Growth'] = df_sorted['Asset Change'] / df_sorted.groupby('conm')['at'].shift()
```

```python
import pandas as pd

sp500_data = pd.read_csv("https://datahub.io/core/s-and-p-500-companies/r/constituents.csv")
sp500_dict = {}

for index, row in sp500_data.iterrows():
    company_name = row['Name']
    ticker_symbol = row['Symbol']
    sp500_dict[company_name] = ticker_symbol

print(sp500_dict)
```

```python
filtered_data = df[df['conm'].isin(sp500_dict)]
print(filtered_data)
```

```python
from fuzzywuzzy import process

def fuzzy_match(name):
    match, score = process.extractOne(name, sp500_dict.keys())
    if score >= 90:  # You can adjust this threshold as needed
        return sp500_dict[match]
    else:
        return None

# Apply fuzzy matching to 'conm' column
df['ticker'] = df['conm'].apply(fuzzy_match)

# Filter out rows where fuzzy matching didn't find a good match
filtered_data = df.dropna()

# Print the filtered data 
print(filtered_data)
```
